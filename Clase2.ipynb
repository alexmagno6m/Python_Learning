# Tipos de Datos en Python
Hasta el momento hemos visto dos tipos de datos:
- (enteros), los que corresponden a numeros naturales: 0, 15, 6000 y sus negativos: -4, -3, -10
- Texto: son caracteres escritos entre comillas: “nombre”, “edad”, “año 2014”, el nombre real es cadena. Mas adelante se presenta una definición mas formal de este tipo de dato.
> ## Tipo de dato Float
> El operador float sirve para almacenar numeros en variables (generalmente cuando se le preguntan al usuario) pero a diferencia de int, permite el uso de números con decimales: 1.4235, 2.5, -4.3.
> ## Tipo de dato: cadena (string)
>Las cadenas son datos de tipo texto, compuestas por caracteres bien sea signos, símbolos, números y alfabetos. Por ejemplo: “fruta”, “mundial 2014”, “grado 11”. Una cadena se escribe entre comillas dobles.
>> ### Composición de una cadena
>> Toda cadena esta compuesta por una longitud, un índice y un(os) elemento(s).
>> - __Longitud__: es el numero de elementos que tienen la cadena.
>> - __Indice__: es la posicion que ocupa un determinado elemente dentro de la cadena
>> - __Elemento__: es el carácter que se encuentra en un determinado índice en la cadena.
