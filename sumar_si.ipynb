# Funcion SUMAR.SI

## Funcionamiento
>
> La funcion SUMAR.Si trabaja de la misma manera que la funcion PROMEDIO.SI, lo unico realmente distinto es la operacion a realizar, como su propio nombre lo dice, en lugar de realizar un promedio efectua una suma condicionada a unos datos especificos (campo __rango suma__) que afectara unicamente a los valores (campo __rango__) que cumplan con la restriccion (campo __criterio__), aspectos a tener en cuenta:
>
> 
>
> - La funcion sumar.si se encuentra en Excel, en la categoria de funciones __Matematicas y Trigonometricas__.
> - Los campos __rango__ y __criterio__ deben pertenecer a la misma columna.
> - El campo __rango suma__ puede pertenecer a una columna diferente.
>
> Si desea obtener mas informacion sobre el funcionamiento de PROMEDIO.SI, siga este [Enlace](https://nbviewer.org/github/alexmagno6m/Informatica_10/blob/main/promedio_si.ipynb)

# Funcion MAX y MIN
## Concepto
> Pertenecen al grupo de funciones estadisticas:
> - La funcion __MAX__ devuelve el numero mas alto (o mayor) de un rango o grupo de rangos
> - La funcion __MIN__ devuelve el numero mas bajo (o menor) de un rango o grupo de rangos
> Para acceder a ella se hace clic en la flecha de la seccion de autosuma donde aparecen como predeterminadas para usar la sintaxis directamente en la celda, o tamien haciendo clic en _Mas funciones_ y seleccionando posteriormente la categoria de estadisticas, donde se presenta la ventana de edicion:
>
> 
>
> Si no se indica lo contrario se usa solo el campo __NUmero 1__
> Alli se selecciona la columna o fila de donde se desea obtener al valor maximo o minimo.
