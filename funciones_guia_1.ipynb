## Definiendo una funcion

Una función que verifica el acceso a una plataforma:

```python
def validar_acceso(usuario, password, rol):
    if usuario == "admin" and password == "1234" and rol == "profesor":
        return "Acceso concedido"
    return "Acceso denegado"
```

> Una función puede multiples sentencias `return` siempre que se encuentren en caminos de ejecución diferentes, por eejemplo en una estructura condicional como en el caso anterior.

> Una función se mantiene inactiva hasta cuando sea llamada en el codigo, una vez invocada se ejecuta, normalmente puede entregar valores de retornos utiles para el contexto del programa.

___

Una función que formatea un mensaje para correo electronico de control de tareas:

```python
def crear_mensaje(nombre, materia, tarea):
    mensaje = f"Hola {nombre}, recuerda entregar la tarea de {materia}: {tarea}."
    return mensaje
```

> Generalmente las variables definidas en el cuerpo de una función son innacesibles al codigo externo a esta. Como en `mensaje` en este caso. Esto se conoce como `local scope` o variables de ambito local.

___

Una función que genera el usuario de correo electronico tomando como entradas los nombres y apellidos:

```python
def generar_correo(nombre, apellido, dominio):
    correo = f"{nombre.lower()}.{apellido.lower()}@{dominio}"
    return correo
```

> Las funciones son reutilizables, ayudan a organizar y mejorar la legibilidad del codigo, siempre que esten bien diseñadas y se acompañen de buenas prácticas.
