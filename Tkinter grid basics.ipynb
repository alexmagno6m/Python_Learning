# Tkinter Grid Basics

## Disposicion de cuadricula

Los argumentos ```column``` y ```row``` permiten posicionar un widget en una posicion especifica dentro de la cuadricula

Por supuesto tambien es posible expandir un widget a lo ancho o largo dentro de las celdas adyacentes, usando los argumentos ```columnspan``` y ```rowspan``` respectivamente.

## Argumento ```sticky```

Por defecto una celda es tan grande como el widget que contiene y lo ubica en centro de la celda horizontal y verticalmente:

Para cambiar el comportamiento original se usa el argumento ```sticky```, el cual especifica a cual borde de la celda debe "pegarse". Las opciones son:

| Sticky | Description                                                                              |
|--------|------------------------------------------------------------------------------------------|
| N      | North or Top Center                                                                      |
| S      | South or Bottom Center                                                                   |
| E      | East or Right Center                                                                     |
| W      | West or Left Center                                                                      |
| NW     | North West or Top Left                                                                   |
| NE     | North East or Top Right                                                                  |
| SE     | South East or Bottom Right                                                               |
| SW     | South West or Bottom Left                                                                |
| NS     | NS stretches the widget vertically. However, it leaves the widget centered horizontally. |
| EW     | EW stretches the widget horizontally. However, it leaves the widget centered vertically. |

Por ejemplo si se usa como valor ```NW``` el widget se posiciona en la esquina superior izquierda de la celda:

Se pueden combinar los valores, ```NS``` por ejemplo, estira el widget verticalmente. Sin embargo, deja el widget centrado horizontalmente:

## Argumento Padding

Las opciones ```padx``` y ```pady``` agrega un margen entre celdas. En cambio los argumentos ```ipadx``` y ```ipady``` sirven para generar un margen interno en esa celda.
