# Color Picker

Color Picker es una app que se emplea en el diseño grafico para poder obtener el codigo hexadecimal de un color de muestra. Es util para tener referencias en el desarrollo de logos o interfaces graficas.

![picker](picker.png)

Pyhton ya trae una libreria lista para diseñar este tipo de apps, llamada: `colorchooser`, la cual proporciona los elementos minimos para que la interfaz sea funcional.

1. Cree un nuevo archivo en su carpeta de trabajo llamado: `color_piocker.py`

2. Copie el siguiente codigo:

```python
from tkinter import *
from tkinter import colorchooser

ventana = Tk()
ventana.title("Color Picker App")
ventana.geometry("300x150")

# Botón para elegir el color
boton = Button(ventana, text="Seleccionar color")
boton.pack(pady=20)

# Etiqueta para mostrar el color elegido
color_label = Label(ventana, text="Color elegido: Ninguno")
color_label.pack(pady=10)

ventana.mainloop()
```

3. Antes de la linea donde aparece la sentencia `ventana = Tk()`, copie el siguiente codigo:

```python
def elegir_color():
    color = colorchooser.askcolor(title="Elige un color")
    if color[1]:
        color_label.config(text=f"Color elegido: {color[1]}", bg=color[1])
```
